```python
# 1. 필요한 라이브러리 임포트
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# 버전 확인 (설치 확인용)
import sklearn
print(f"Scikit-learn version: {sklearn.__version__}")

# 2. 예제 데이터 로드 (Iris 데이터셋)
iris = load_iris()
X = iris.data  # type: ignore # 특징 데이터 (꽃받침 길이 등)
y = iris.target # type: ignore # 정답 레이블 (붓꽃 종류)

# 데이터프레임으로 변환하여 상위 5개 행 확인
df = pd.DataFrame(X, columns=iris.feature_names) # type: ignore
df['species'] = y
print("\n[데이터 미리보기]")
print(df.head())
# 5.1 3.5 1.4 0.2
# 3. 학습용(Train)과 테스트용(Test) 데이터 분리 (8:2 비율)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.6, random_state=42)

# 4. 모델 생성 및 학습 (의사결정나무 모델 사용)
model = DecisionTreeClassifier()
model.fit(X_train, y_train)

# 5. 예측 수행
y_pred = model.predict(X_test)

# 6. 정확도 평가
accuracy = accuracy_score(y_test, y_pred)
print(f"\n모델 정확도: {accuracy:.2f}")
```
